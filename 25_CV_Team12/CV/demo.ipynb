def color_transfer

```python
def get_mean_and_std(x):
  x_mean, x_std = cv2.meanStdDev(x)
  x_mean = np.hstack(np.around(x_mean,2))
  x_std = np.hstack(np.around(x_std,2))
  return x_mean, x_std

def color_transfer():
  s = cv2.imread(SOURCE_PATH)
  t = cv2.imread(TARGET_PATH)

  s_lab = cv2.cvtColor(s.copy(), cv2.COLOR_BGR2LAB) # 복사본 사용
  t_lab = cv2.cvtColor(t.copy(), cv2.COLOR_BGR2LAB) # 복사본 사용

  s_mean, s_std = get_mean_and_std(s_lab)
  t_mean, t_std = get_mean_and_std(t_lab)

  height, width, channel = s_lab.shape
  for i in range(height):
    for j in range(width):
      for k in range(channel):
        x = s_lab[i,j,k]

        if s_std[k] == 0:
            x_new = t_mean[k]
        else:
            x_new = ((x-s_mean[k])*(t_std[k]/s_std[k]))+t_mean[k]

        x_new = round(x_new)
        x_new = 0 if x_new<0 else x_new
        x_new = 255 if x_new>255 else x_new
        s_lab[i,j,k] = x_new

  result_bgr = cv2.cvtColor(s_lab, cv2.COLOR_LAB2BGR)

  print("Color Transfer가 완료되었습니다. 원본 이미지와 결과 이미지를 출력합니다.")
  return s, t, result_bgr
```

def adjust mean luminace - 평균 휘도(Mean Luminance) 일치

```python
def adjust_mean_luminance(result_img_bgr, target_img_bgr):
    """결과 이미지의 L 채널 평균을 타겟 이미지의 L 채널 평균과 일치시킵니다."""
    result_lab = cv2.cvtColor(result_img_bgr, cv2.COLOR_BGR2LAB)
    target_lab = cv2.cvtColor(target_img_bgr, cv2.COLOR_BGR2LAB)

    L_result = result_lab[:, :, 0]
    L_target = target_lab[:, :, 0]

    mean_result = np.mean(L_result)
    mean_target = np.mean(L_target)

    delta_L = mean_target - mean_result
    L_adjusted = L_result.astype(np.float32) + delta_L
    L_adjusted = np.clip(L_adjusted, 0, 255).astype(np.uint8)

    result_lab[:, :, 0] = L_adjusted
    return cv2.cvtColor(result_lab, cv2.COLOR_LAB2BGR)
```

def adjust chroma variance - 색상 채도 (Chroma/Saturation) 조정 (A, B 채널 분산 일치)

```python
def adjust_chroma_variance(result_img_bgr, target_img_bgr, scale_factor=1.0):
    """결과 이미지의 A/B 채널 분산을 타겟 이미지의 분산에 맞게 조정합니다."""
    result_lab = cv2.cvtColor(result_img_bgr, cv2.COLOR_BGR2LAB)
    target_lab = cv2.cvtColor(target_img_bgr, cv2.COLOR_BGR2LAB)

    A_result = result_lab[:, :, 1].astype(np.float32) - 128
    B_result = result_lab[:, :, 2].astype(np.float32) - 128
    A_target = target_lab[:, :, 1].astype(np.float32) - 128
    B_target = target_lab[:, :, 2].astype(np.float32) - 128

    std_A_result = np.std(A_result)
    std_B_result = np.std(B_result)
    std_A_target = np.std(A_target)
    std_B_target = np.std(B_target)

    A_adjusted = A_result * (std_A_target / std_A_result) * scale_factor if std_A_result > 1e-6 else A_result
    B_adjusted = B_result * (std_B_target / std_B_result) * scale_factor if std_B_result > 1e-6 else B_result

    result_lab[:, :, 1] = np.clip(A_adjusted + 128, 0, 255).astype(np.uint8)
    result_lab[:, :, 2] = np.clip(B_adjusted + 128, 0, 255).astype(np.uint8)

    return cv2.cvtColor(result_lab, cv2.COLOR_LAB2BGR)
```

def remove color cast - 색상 편향 (Color Cast) 추정 및 제거


```python
def remove_color_cast(result_img_bgr):
    """결과 이미지의 A/B 채널 평균을 128(중앙값)로 이동시켜 색상 편향을 제거합니다."""
    result_lab = cv2.cvtColor(result_img_bgr, cv2.COLOR_BGR2LAB)

    A_result = result_lab[:, :, 1].astype(np.float32)
    B_result = result_lab[:, :, 2].astype(np.float32)

    mean_A = np.mean(A_result)
    mean_B = np.mean(B_result)

    delta_A = 128 - mean_A
    delta_B = 128 - mean_B

    A_adjusted = A_result + delta_A
    B_adjusted = B_result + delta_B

    result_lab[:, :, 1] = np.clip(A_adjusted, 0, 255).astype(np.uint8)
    result_lab[:, :, 2] = np.clip(B_adjusted, 0, 255).astype(np.uint8)

    return cv2.cvtColor(result_lab, cv2.COLOR_LAB2BGR)
```

def remove color noise - 색상 노이즈 제거 (가우시안 블러 사용)

```python
def remove_color_noise(result_img_bgr, kernel_size=3):
    """A/B 채널에 존재하는 고주파 노이즈를 가우시안 블러를 사용하여 제거합니다."""
    result_lab = cv2.cvtColor(result_img_bgr, cv2.COLOR_BGR2LAB)

    L, A, B = cv2.split(result_lab)

    A_denoised = cv2.GaussianBlur(A, (kernel_size, kernel_size), 0)
    B_denoised = cv2.GaussianBlur(B, (kernel_size, kernel_size), 0)

    result_lab_denoised = cv2.merge([L, A_denoised, B_denoised])

    return cv2.cvtColor(result_lab_denoised, cv2.COLOR_LAB2BGR)
```

def adjust white black points

```python
def adjust_white_black_points(source_bgr, target_bgr, shadow_clip=0.01, highlight_clip=0.01):
    """

      Match Image의 L 채널 분포를 분석하여 Black/White Point를 추출하고 Source Image에 적용합니다.

      :param source_bgr: Color Transfer가 적용될 Input Image (BGR)

      :param target_bgr: 스타일을 추출할 Match Image (BGR)

      :param shadow_clip: 무시할 가장 어두운 픽셀의 비율 (예: 1% = 0.01)

      :param highlight_clip: 무시할 가장 밝은 픽셀의 비율 (예: 1% = 0.01)

      :return: 톤 포인트가 조정된 BGR 이미지

    """
    s_lab = cv2.cvtColor(source_bgr.copy(), cv2.COLOR_BGR2LAB)
    t_lab = cv2.cvtColor(target_bgr.copy(), cv2.COLOR_BGR2LAB)

    s_L = s_lab[:, :, 0]
    t_L = t_lab[:, :, 0]

    hist, bins = np.histogram(t_L.flatten(), 256, [0, 256])
    total_pixels = t_L.size
    cdf = hist.cumsum()

    clip_pixels_shadow = total_pixels * shadow_clip
    black_point_target = np.searchsorted(cdf, clip_pixels_shadow)

    clip_pixels_highlight = total_pixels * (1 - highlight_clip)
    white_point_target = np.searchsorted(cdf, clip_pixels_highlight)

    print(f"  [Tone Point 추출] Black Point (L): {black_point_target}, White Point (L): {white_point_target}")

    s_L_float = s_L.astype(np.float32)
    L_range = white_point_target - black_point_target

    if L_range <= 0:
        print("  [경고] Target Image의 Black/White Point가 너무 가까워 조정 스킵.")
        return source_bgr

    s_L_new = (s_L_float - black_point_target) * (255.0 / L_range)
    s_L_new = np.clip(s_L_new, 0, 255).astype(np.uint8)
    s_lab[:, :, 0] = s_L_new
    result_bgr = cv2.cvtColor(s_lab, cv2.COLOR_LAB2BGR)

    return result_bgr
```

def display images cv2 imshow

```python
def display_images_cv2_imshow(s_original, t_original, final_color_transfer_result, *other_results):
    """
    cv2_imshow를 사용하여 여러 BGR 이미지를 순차적으로 출력합니다.
    """

    images_to_show = [
        (s_original, "1. 원본 Source Image"),
        (t_original, "2. Target Image (스타일 참조)"),
        (final_color_transfer_result, "3. Color Transfer 결과"),
    ]

    current_step = 4
    for img, title in other_results:
        images_to_show.append((img, f"{current_step}. {title}"))
        current_step += 1

    print("\n--- 이미지 출력 시작 (cv2_imshow 사용) ---")

    for img, title in images_to_show:
        if img is not None:
            print(f"\n--- {title} ---")
            cv2_imshow(img)
        else:
            print(f"\n--- {title} --- (이미지 없음/None)")
```

def save images to folder

```python
def save_images_to_folder(image_list_with_names, output_folder):
    """
    BGR 이미지 목록을 지정된 폴더에 저장합니다.
    Args:
        image_list_with_names (list): [(image_data, file_name), ...] 형태의 리스트
        output_folder (str): 이미지를 저장할 폴더 경로
    """
    os.makedirs(output_folder, exist_ok=True)

    print(f"\n--- 결과 이미지 저장 시작: {output_folder} ---")

    count = 0
    for img, filename in image_list_with_names:
        if img is not None:
            save_path = os.path.join(output_folder, filename)
            cv2.imwrite(save_path, img)
            print(f"저장 완료: {filename}")
            count += 1
        else:
            print(f"저장 건너뜀: {filename} (이미지 데이터 없음)")

    print(f"---  총 {count}개 이미지 저장 완료 ---")
```

image->RGB 변환

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def load_rgb(path, max_side=1200):
    img = Image.open(path).convert("RGB")
    w,h = img.size
    s = max(w,h)
    if s > max_side:
        scale = max_side / s
        img = img.resize((int(w*scale), int(h*scale)), Image.LANCZOS)
    return img

def show(img, title=""):
    plt.figure(figsize=(6,6))
    plt.imshow(img)
    plt.axis("off")
    if title: plt.title(title)
    plt.show()
```

RGB-> LAB 변환 & LAB->RGB 복원

```python
#0~255->0~1정규화
def rgb_to_xyz(rgb):
    rgb = rgb / 255.0
    mask = rgb > 0.04045
    rgb_lin = np.where(mask, ((rgb + 0.055)/1.055)**2.4, rgb/12.92)

    M = np.array([[0.4124564, 0.3575761, 0.1804375],
                  [0.2126729, 0.7151522, 0.0721750],
                  [0.0193339, 0.1191920, 0.9503041]], dtype=np.float32)
    return rgb_lin @ M.T

#XYZ를 D65 white으로 정규화
def xyz_to_lab(xyz):
    # D65 white
    white = np.array([0.95047, 1.0, 1.08883], dtype=np.float32)
    x = xyz[...,0] / white[0]
    y = xyz[...,1] / white[1]
    z = xyz[...,2] / white[2]

    eps = 216/24389
    k = 24389/27

    def f(t):
        return np.where(t > eps, np.cbrt(t), (k*t + 16)/116)

    fx, fy, fz = f(x), f(y), f(z)
    L = 116*fy - 16
    a = 500*(fx - fy)
    b = 200*(fy - fz)
    return np.stack([L,a,b], axis=-1)

def lab_to_xyz(lab):
    white = np.array([0.95047, 1.0, 1.08883], dtype=np.float32)
    L,a,b = lab[...,0], lab[...,1], lab[...,2]
    fy = (L + 16)/116
    fx = fy + a/500
    fz = fy - b/200

    eps = 216/24389
    k = 24389/27

    def finv(t):
        t3 = t**3
        return np.where(t3 > eps, t3, (116*t - 16)/k)

    x = finv(fx) * white[0]
    y = finv(fy) * white[1]
    z = finv(fz) * white[2]
    return np.stack([x,y,z], axis=-1)

def xyz_to_rgb(xyz):
    M = np.array([[ 3.2404542, -1.5371385, -0.4985314],
                  [-0.9692660,  1.8760108,  0.0415560],
                  [ 0.0556434, -0.2040259,  1.0572252]], dtype=np.float32)
    rgb_lin = xyz @ M.T
    rgb_lin = np.clip(rgb_lin, 0, 1)


    mask = rgb_lin > 0.0031308
    rgb = np.where(mask, 1.055*(rgb_lin**(1/2.4)) - 0.055, 12.92*rgb_lin)
    rgb = np.clip(rgb, 0, 1)
    return (rgb * 255.0)

def rgb_to_lab(img_np):
    xyz = rgb_to_xyz(img_np.astype(np.float32))
    return xyz_to_lab(xyz)

def lab_to_rgb(lab):
    xyz = lab_to_xyz(lab.astype(np.float32))
    return xyz_to_rgb(xyz)

def reinhard_transfer(src_rgb, tgt_rgb, eps=1e-6):
    # src -> input, tgt -> target
    src_lab = rgb_to_lab(src_rgb)
    tgt_lab = rgb_to_lab(tgt_rgb)

    out = src_lab.copy()
    for c in range(3):
        s_mean, s_std = src_lab[...,c].mean(), src_lab[...,c].std() + eps
        t_mean, t_std = tgt_lab[...,c].mean(), tgt_lab[...,c].std() + eps
        out[...,c] = (out[...,c] - s_mean) * (t_std / s_std) + t_mean

    out_rgb = lab_to_rgb(out)
    return np.clip(out_rgb, 0, 255).astype(np.uint8)
```

```python
def match_hist_1d(src, tgt, bins=256):

    src_hist, _ = np.histogram(src, bins=bins, range=(0,255), density=True)
    tgt_hist, _ = np.histogram(tgt, bins=bins, range=(0,255), density=True)

    src_cdf = np.cumsum(src_hist); src_cdf /= src_cdf[-1]
    tgt_cdf = np.cumsum(tgt_hist); tgt_cdf /= tgt_cdf[-1]

    mapping = np.interp(src_cdf, tgt_cdf, np.arange(bins))
    src_vals = np.clip(src.astype(np.int32), 0, bins-1)
    return mapping[src_vals].astype(np.uint8)

def rgb_to_y(rgb):
    rgb = rgb.astype(np.float32)
    y = 0.2126*rgb[...,0] + 0.7152*rgb[...,1] + 0.0722*rgb[...,2]
    return np.clip(y, 0, 255).astype(np.uint8)

def apply_luma_match(src_rgb, tgt_rgb):
    src = src_rgb.copy()
    y_src = rgb_to_y(src)
    y_tgt = rgb_to_y(tgt_rgb)

    y_matched = match_hist_1d(y_src.flatten(), y_tgt.flatten()).reshape(y_src.shape)

    y_src_f = y_src.astype(np.float32) + 1e-6
    ratio = (y_matched.astype(np.float32) + 1e-6) / y_src_f
    out = np.clip(src.astype(np.float32) * ratio[...,None], 0, 255).astype(np.uint8)
    return out
```

```python
def reinhard_ab_only(src_rgb, tgt_rgb, eps=1e-6):
    src_lab = rgb_to_lab(src_rgb)
    tgt_lab = rgb_to_lab(tgt_rgb)

    out = src_lab.copy()
    for c in [1, 2]:  # a,b만
        s_mean, s_std = src_lab[...,c].mean(), src_lab[...,c].std() + eps
        t_mean, t_std = tgt_lab[...,c].mean(), tgt_lab[...,c].std() + eps
        out[...,c] = (out[...,c] - s_mean) * (t_std / s_std) + t_mean

    out_rgb = lab_to_rgb(out)
    return np.clip(out_rgb, 0, 255).astype(np.uint8)

step1 = reinhard_ab_only(inp_np, target_np)
show(Image.fromarray(step1), "AB-only color transfer")
```

```python
import numpy as np
from PIL import Image, ImageFilter
import cv2
from google.colab.patches import cv2_imshow

def show_np(arr, title=""):
    img = np.clip(arr,0,255).astype(np.uint8)

    if img.ndim == 3 and img.shape[2] == 3:
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    if title:
        print(title)
    cv2_imshow(img)

def to_img(arr):
    return Image.fromarray(np.clip(arr,0,255).astype(np.uint8))


#밝기 계산
def luma(rgb):
    rgb = rgb.astype(np.float32)
    return 0.2126*rgb[...,0] + 0.7152*rgb[...,1] + 0.0722*rgb[...,2]

def pct(x,p):
    return float(np.percentile(x,p))

#RGB<->HSV
def rgb_to_hsv(rgb):
    x = np.clip(rgb/255.0, 0, 1).astype(np.float32)
    r,g,b = x[...,0], x[...,1], x[...,2]
    mx = np.max(x, axis=-1)
    mn = np.min(x, axis=-1)
    diff = mx - mn

    h = np.zeros_like(mx)
    s = np.zeros_like(mx)
    v = mx

    s = np.where(mx == 0, 0, diff / (mx + 1e-6))

    # hue
    mask = diff > 1e-6
    rc = ((mx - r) / (diff + 1e-6))
    gc = ((mx - g) / (diff + 1e-6))
    bc = ((mx - b) / (diff + 1e-6))

    h = np.where((mx==r) & mask, (bc - gc), h)
    h = np.where((mx==g) & mask, 2.0 + (rc - bc), h)
    h = np.where((mx==b) & mask, 4.0 + (gc - rc), h)
    h = (h/6.0) % 1.0
    return np.stack([h,s,v], axis=-1)

def hsv_to_rgb(hsv):
    h,s,v = hsv[...,0], hsv[...,1], hsv[...,2]
    i = np.floor(h*6.0).astype(np.int32)
    f = h*6.0 - i
    p = v*(1.0 - s)
    q = v*(1.0 - f*s)
    t = v*(1.0 - (1.0 - f)*s)
    i = i % 6

    r = np.select([i==0,i==1,i==2,i==3,i==4,i==5], [v,q,p,p,t,v], default=v)
    g = np.select([i==0,i==1,i==2,i==3,i==4,i==5], [t,v,v,q,p,p], default=v)
    b = np.select([i==0,i==1,i==2,i==3,i==4,i==5], [p,p,t,v,v,q], default=v)
    rgb = np.stack([r,g,b], axis=-1)
    return np.clip(rgb*255.0, 0, 255).astype(np.float32)

#밝기 구간 mask(hightlights/shadows 조절)
def soft_mask_by_luma(Y, center, width):

    return np.clip(1.0 - np.abs(Y - center)/max(width,1e-6), 0, 1)

# target & base 에서 수치 뽑기
def extract_tone_stats(rgb_np):
    Y = luma(rgb_np).flatten()
    return {
        "p1":  pct(Y, 1),
        "p50": pct(Y, 50),
        "p99": pct(Y, 99),
        "p10": pct(Y, 10),
        "p90": pct(Y, 90),
        "spread_p90_p10": pct(Y,90) - pct(Y,10),
        "meanY": float(Y.mean())
    }

#clarity: RGB를 흑백으로 변환하고 중간 크기 구조만 남김
def extract_clarity_strength(rgb_np):
    imgL = to_img(rgb_np).convert("L")
    b1 = np.array(imgL.filter(ImageFilter.GaussianBlur(radius=1.0))).astype(np.float32)/255.0
    b2 = np.array(imgL.filter(ImageFilter.GaussianBlur(radius=6.0))).astype(np.float32)/255.0
    dog = b1 - b2
    return {"dog_std": float(dog.std() + 1e-6)}

#sharpness edge 강도 추정
def extract_sharpness(rgb_np):
    imgL = to_img(rgb_np).convert("L")
    edges = imgL.filter(ImageFilter.FIND_EDGES)
    e = np.array(edges).astype(np.float32)/255.0
    return {"edge_std": float(e.std() + 1e-6)}

def extract_color_stats(rgb_np):
    hsv = rgb_to_hsv(rgb_np.astype(np.float32))
    S = hsv[...,1].flatten()
    V = hsv[...,2].flatten()

    # temperature/tint는 RGB 평균으로 근사
    mean_rgb = rgb_np.reshape(-1,3).astype(np.float32).mean(axis=0)
    return {
        "meanS": float(S.mean()),
        "p90S": float(np.percentile(S, 90)),
        "meanV": float(V.mean()),
        "meanR": float(mean_rgb[0]),
        "meanG": float(mean_rgb[1]),
        "meanB": float(mean_rgb[2]),
    }

#base를 타겟처럼 맞추기
def blend(base, y, strength):
    return np.clip((1-strength)*base + strength*y, 0, 255)

#블랙/화이트 레벨 맞추기
def match_levels(base_np, base_tone, tgt_tone, strength=0.7):
    x = base_np.astype(np.float32)
    b1, b99 = base_tone["p1"], base_tone["p99"]
    t1, t99 = tgt_tone["p1"], tgt_tone["p99"]

    y = (x - b1) * (255.0 / max(b99 - b1, 1e-6))
    y = np.clip(y, 0, 255)
    y = y * (max(t99 - t1, 1e-6) / 255.0) + t1
    y = np.clip(y, 0, 255)
    return blend(x, y, strength)
#노출
def match_exposure(base_np, tgt_tone, strength=0.6):
    x = base_np.astype(np.float32)
    Y = luma(x)/255.0
    src = float(np.mean(Y))
    tgt = float(np.clip(tgt_tone["meanY"]/255.0, 1e-4, 1.0))
    src = float(np.clip(src, 1e-4, 1.0))
    gain = tgt/src
    gain = float(np.clip(gain, 0.6, 1.6))
    y = np.clip(x * gain, 0, 255)
    return blend(x, y, strength), gain

#midetone
def match_midtone_gamma(base_np, tgt_tone, strength=0.5):
    x = base_np.astype(np.float32)
    Y = luma(x)/255.0
    src_med = float(np.median(Y))
    tgt_med = float(np.clip(tgt_tone["p50"]/255.0, 1e-4, 1.0))
    src_med = float(np.clip(src_med, 1e-4, 1.0))
    gamma = np.log(tgt_med) / np.log(src_med)
    gamma = float(np.clip(gamma, 0.7, 1.4))
    Yg = np.clip(Y**gamma, 0, 1)
    ratio = (Yg / np.clip(Y, 1e-4, 1.0))[...,None]
    y = np.clip(x * ratio, 0, 255)
    return blend(x, y, strength), gamma

#s-curve 대비
def apply_s_curve(Y, amount):

    a = float(np.clip(amount, -1.0, 1.0))
    k = 1.0 + 4.0*abs(a)
    Y2 = 0.5 + 0.5*np.tanh(k*(Y-0.5)) / np.tanh(0.5*k)
    return (1-abs(a))*Y + abs(a)*Y2 if a>=0 else (1-abs(a))*Y + abs(a)*(1-(0.5 + 0.5*np.tanh(k*(1-Y-0.5))/np.tanh(0.5*k)))
#대비
def add_contrast(base_np, amount=1.10):

    x = base_np.astype(np.float32)
    mean = x.mean(axis=(0,1), keepdims=True)
    out = (x - mean) * amount + mean
    return np.clip(out, 0, 255)

def match_contrast(base_np, base_tone, tgt_tone, strength=0.6):
    x = base_np.astype(np.float32)
    cur = float(base_tone["spread_p90_p10"]/255.0)
    tgt = float(tgt_tone["spread_p90_p10"]/255.0)
    ratio = float(np.clip(tgt/(cur+1e-6), 0.7, 1.4))

    amount = float(np.clip((ratio-1.0)*1.2, -0.8, 0.8))
    amount *= strength

    Y = luma(x)/255.0
    Y2 = np.clip(apply_s_curve(Y, amount), 0, 1)
    ratioY = (Y2 / np.clip(Y, 1e-4, 1.0))[...,None]
    y = np.clip(x * ratioY, 0, 255)
    return blend(x, y, 1.0), amount

def match_highlights_shadows(base_np, tgt_tone, strength=0.6):
    x = base_np.astype(np.float32)
    Y = luma(x)/255.0

    # target의 highlight/shadow 근사
    t_hi = float(tgt_tone["p90"]/255.0)
    t_sh = float(tgt_tone["p10"]/255.0)

    cur_hi = float(np.percentile(Y, 90))
    cur_sh = float(np.percentile(Y, 10))

    m_hi = soft_mask_by_luma(Y, center=0.85, width=0.20)
    m_sh = soft_mask_by_luma(Y, center=0.15, width=0.20)

    d_hi = float(np.clip(t_hi - cur_hi, -0.15, 0.15)) * strength
    d_sh = float(np.clip(t_sh - cur_sh, -0.15, 0.15)) * strength

    Y2 = np.clip(Y + m_hi*d_hi + m_sh*d_sh, 0, 1)
    ratioY = (Y2 / np.clip(Y, 1e-4, 1.0))[...,None]
    y = np.clip(x * ratioY, 0, 255)
    return y, d_hi, d_sh

def match_saturation_vibrance(base_np, base_col, tgt_col, strength=0.6):
    x = base_np.astype(np.float32)
    hsv = rgb_to_hsv(x)
    S = hsv[...,1]
    V = hsv[...,2]

    curS = float(base_col["meanS"])
    tgtS = float(tgt_col["meanS"])
    ratio = float(np.clip(tgtS/(curS+1e-6), 0.7, 1.5))

    # vibrance 근사
    vib_mask = np.clip(1.0 - S, 0, 1)
    sat_gain = 1.0 + (ratio-1.0)*strength
    vib_gain = 1.0 + (ratio-1.0)*strength*0.8

    S2 = np.clip(S * sat_gain * (1.0 + vib_mask*(vib_gain-1.0)), 0, 1)
    hsv2 = hsv.copy()
    hsv2[...,1] = S2
    y = hsv_to_rgb(hsv2)
    return blend(x, y, 1.0), sat_gain

def match_temperature_tint(base_np, base_col, tgt_col, strength=0.6):
    x = base_np.astype(np.float32)

    src_rb = float(base_col["meanR"] - base_col["meanB"])
    tgt_rb = float(tgt_col["meanR"] - tgt_col["meanB"])
    src_g  = float(base_col["meanG"])
    tgt_g  = float(tgt_col["meanG"])

    d_temp = float(np.clip((tgt_rb - src_rb)/255.0, -0.12, 0.12)) * strength
    d_tint = float(np.clip((tgt_g  - src_g )/255.0, -0.08, 0.08)) * strength

    y = x.copy()
    y[...,0] = np.clip(y[...,0] * (1.0 + d_temp), 0, 255)
    y[...,2] = np.clip(y[...,2] * (1.0 - d_temp), 0, 255)
    y[...,1] = np.clip(y[...,1] * (1.0 + d_tint), 0, 255)
    return blend(x, y, 1.0), d_temp, d_tint

def match_clarity(base_np, base_cl, tgt_cl, strength=0.7):
    cur = base_cl["dog_std"]
    tgt = tgt_cl["dog_std"]
    ratio = float(np.clip(tgt/cur, 0.7, 1.6))
    amount = float(np.clip((ratio - 1.0) * 0.35, -0.25, 0.25)) * strength

    imgL = to_img(base_np).convert("L")
    y  = np.array(imgL).astype(np.float32)/255.0
    b1 = np.array(imgL.filter(ImageFilter.GaussianBlur(radius=1.0))).astype(np.float32)/255.0
    b2 = np.array(imgL.filter(ImageFilter.GaussianBlur(radius=6.0))).astype(np.float32)/255.0
    dog = b1 - b2
    y2 = np.clip(y + amount*dog, 0, 1)

    ratioY = (y2 / (y + 1e-6))[...,None]
    out = np.clip(base_np.astype(np.float32) * ratioY, 0, 255)
    return out, amount

def match_unsharp(base_np, base_sh, tgt_sh, strength=0.7, radius=1.2, threshold=2):
    cur = base_sh["edge_std"]
    tgt = tgt_sh["edge_std"]
    ratio = float(np.clip(tgt/cur, 0.7, 1.6))

    percent = 120 * ratio
    percent = 120 + (percent - 120) * strength
    percent = int(np.clip(percent, 60, 180))

    img = to_img(base_np)
    out = img.filter(ImageFilter.UnsharpMask(radius=radius, percent=percent, threshold=threshold))
    return np.array(out).astype(np.float32), percent

def match_denoise(base_np, tgt_strength, strength=0.6):

    x = base_np.astype(np.float32)
    r = float(np.clip(tgt_strength, 0.0, 1.0)) * strength
    radius = 0.2 + 1.2*r
    y = np.array(to_img(x).filter(ImageFilter.GaussianBlur(radius=radius))).astype(np.float32)
    return blend(x, y, strength), radius

def apply_vignette(base_np, tgt_vignette=0.2, strength=0.6):
    x = base_np.astype(np.float32)
    h,w = x.shape[:2]
    yy,xx = np.mgrid[0:h,0:w].astype(np.float32)
    cy,cx = (h-1)/2.0, (w-1)/2.0
    r = np.sqrt(((yy-cy)/cy)**2 + ((xx-cx)/cx)**2)
    mask = np.clip(1.0 - tgt_vignette*(r**1.6), 0.6, 1.0)
    y = np.clip(x * mask[...,None], 0, 255)
    return blend(x, y, strength)

#base->target match
base = ab_uns.astype(np.float32)
tgt  = target_np.astype(np.float32)

# extract target & base stats
t_tone = extract_tone_stats(tgt)
b_tone = extract_tone_stats(base)
t_cl   = extract_clarity_strength(tgt)
b_cl   = extract_clarity_strength(base)
t_sh   = extract_sharpness(tgt)
b_sh   = extract_sharpness(base)
t_col  = extract_color_stats(tgt)
b_col  = extract_color_stats(base)

print("=== TARGET tone ===", t_tone)
print("=== BASE   tone ===", b_tone)
print("=== TARGET clarity/sharp ===", t_cl, t_sh)
print("=== BASE   clarity/sharp ===", b_cl, b_sh)
print("=== TARGET color ===", t_col)
print("=== BASE   color ===", b_col)

# exposure
x0, used_gain = match_exposure(base, t_tone, strength=0.35)

# blacks/whites via levels
x1 = match_levels(base, extract_tone_stats(x0), t_tone, strength=0.65)

# contrast (S-curve)
xC, used_contrast = match_contrast(base, extract_tone_stats(x1), t_tone, strength=0.60)

# highlights/shadows
xHS, dh, ds = match_highlights_shadows(base, t_tone, strength=0.55)

# midtone gamma
x2, used_gamma = match_midtone_gamma(base, t_tone, strength=0.35)

# temperature/tint (WB)
xWB, dtemp, dtint = match_temperature_tint(base, extract_color_stats(x2), t_col, strength=0.60)

# saturation/vibrance
xSat, sat_gain = match_saturation_vibrance(base, extract_color_stats(xWB), t_col, strength=0.60)

# clarity/texture-ish
x3, used_amount = match_clarity(base, extract_clarity_strength(xSat), t_cl, strength=0.70)

# sharpen
x4, used_percent = match_unsharp(base, extract_sharpness(x3), t_sh, strength=0.70)

# mild denoise (optional)
softness = float(np.clip((b_sh["edge_std"] - t_sh["edge_std"]) / (b_sh["edge_std"] + 1e-6), 0, 1))
xDN, dn_radius = match_denoise(x4, softness, strength=0.35)

# vignette (optional)
xVG = apply_vignette(xDN, tgt_vignette=0.18, strength=0.35)
import os
import re

def sanitize_filename(name):
    name = name.strip()
    name = re.sub(r"[^\w\-_\. ]", "", name)
    name = name.replace(" ", "_")
    return name

def save_np(arr, title, out_dir="outputs"):
    os.makedirs(out_dir, exist_ok=True)
    img = np.clip(arr, 0, 255).astype(np.uint8)
    img = Image.fromarray(img)
    fname = sanitize_filename(title) + ".png"
    path = os.path.join(out_dir, fname)
    img.save(path)
    print(f"saved: {path}")
def save(arr, title, out_dir="outputs"):
    save_np(arr, title, out_dir)

if not os.path.exists(OUTPUT_SAVE_FOLDER):
    try:
        os.makedirs(OUTPUT_SAVE_FOLDER)
        print(f"출력 폴더 생성 완료: {OUTPUT_SAVE_FOLDER}")
    except Exception as e:
        print(f"폴더 생성 실패: {e}")
else:
    print(f"출력 폴더 확인: {OUTPUT_SAVE_FOLDER}")

# 파일 이름
images_to_save = [
    (base, "b2_Base"),
    (x0,   "b2_Exposure"),
    (x1,   "b2_B_W"),
    (xC,   "b2_Contrast"),
    (xHS,  "b2_H_S"),
    (xWB,  "b2_WB"),
    (xSat, "b2_S_V"),
    (x3,   "b2_Clarity"),
    (x4,   "b2_Sharpen"),
    (xVG,  "b2_d_v"),
]

def save_images_to_folder(image_list, folder_path):
    """
    (이미지 배열, 파일 이름) 튜플 리스트를 지정된 폴더에 저장합니다.
    """
    for arr, title in image_list:
        save(arr, title, out_dir=folder_path)


# 4. 이미지 저장 실행
save_images_to_folder(images_to_save, OUTPUT_SAVE_FOLDER)

print(f"\nUsed exposure_gain={used_gain:.3f}")
print(f"Used contrast_amount={used_contrast:.3f}, hi_delta={dh:.3f}, sh_delta={ds:.3f}")
print(f"Used gamma={used_gamma:.3f}, temp_delta={dtemp:.3f}, tint_delta={dtint:.3f}, sat_gain={sat_gain:.3f}")
print(f"Used clarity_amount={used_amount:.3f}, unsharp_percent={used_percent}, denoise_radius={dn_radius:.3f}")

show_np(base, "BASE (only AB+unsharpen)")
show_np(x0,   f"Exposure gain={used_gain:.3f}")
show_np(x1,   "Blacks/Whites (levels p1/p99)")
show_np(xC,   f"Contrast (S-curve damount={used_contrast:.3f})")
show_np(xHS,  f"Highlights/Shadows (dh={dh:.3f}, ds={ds:.3f})")
show_np(xWB,  f"WB (temp={dtemp:.3f}, tint={dtint:.3f})")
show_np(xSat, f"Saturation/Vibrance (gain={sat_gain:.3f})")
show_np(x3,   f"Clarity (amount={used_amount:.3f})")
show_np(x4,   f"Sharpen (percent={used_percent})")
show_np(xVG,  "denoise +vignette")
```

```python
def calculate_average_scores(file_names):
    """
    여러 CSV 파일에서 'File_Name'별 'Style_Similarity_Score'의 평균을 계산합니다.
    """

    all_data = []
    found_files_count = 0

    for file_name in file_names:
        print(f"--- 파일 로드 시도: {file_name} ---")
        try:
            df = pd.read_csv(file_name)

            if 'File_Name' not in df.columns or 'Style_Similarity_Score' not in df.columns:
                print(f"경고: {file_name} 파일에 필요한 컬럼('File_Name', 'Style_Similarity_Score')이 없습니다. 건너뜁니다.")
                continue

            df = df[~df['File_Name'].isin(['---', 'Average Score', ''])].copy()
            df['Style_Similarity_Score'] = pd.to_numeric(
                df['Style_Similarity_Score'],
                errors='coerce'
            )

            df.dropna(subset=['Style_Similarity_Score'], inplace=True)

            all_data.append(df)
            found_files_count += 1
            print(f"성공: {file_name} 파일의 {len(df)}개 항목을 수집했습니다.")

        except FileNotFoundError:
            print(f"오류: {file_name} 파일을 찾을 수 없습니다. 파일을 업로드해 주세요.")
        except Exception as e:
            print(f"오류: {file_name} 파일 처리 중 예상치 못한 오류 발생: {e}")

    if found_files_count == 0:
        print("\n오류: 처리할 유효한 데이터가 없어 평균 계산을 중단합니다.")
        return pd.DataFrame()
    elif found_files_count < len(file_names):
        print(f"\n경고: 요청하신 {len(file_names)}개 중 {found_files_count}개의 파일만 사용하여 평균을 계산합니다.")

    combined_df = pd.concat(all_data, ignore_index=True)
    average_scores_df = combined_df.groupby('File_Name')['Style_Similarity_Score'].mean().reset_index()
    average_scores_df.columns = ['File_Name', 'Average_Similarity_Score']
    average_scores_df['Average_Similarity_Score'] = average_scores_df['Average_Similarity_Score'].apply(lambda x: f'{x:.4f}')
    average_scores_df.sort_values(by='File_Name', inplace=True)

    return average_scores_df

FILE_NAMES = [
    "style_osaka_output.csv",
    # "style_camera_output.csv",
    "style_flower_output.csv",
    "style_yunah_output.csv"
]

OUTPUT_FILE_NAME = "average_style_similarity.csv"

final_average_df = calculate_average_scores(FILE_NAMES)

if not final_average_df.empty:
    final_average_df.to_csv(OUTPUT_FILE_NAME, index=False)
    print(f"최종 평균 CSV 파일 저장 완료: {OUTPUT_FILE_NAME}")

    print("\n[최종 평균 데이터 (상위 5개 항목)]")
    print(final_average_df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
import pandas as pd
import numpy as np
import os

def calculate_average_scores(file_names):
    """
    여러 CSV 파일에서 'File_Name'별 'Content_Similarity_Score'의 평균을 계산합니다.
    """

    SCORE_COLUMN = 'Content_Similarity_Score'
    FILE_COLUMN = 'File_Name'

    all_data = []
    found_files_count = 0

    for file_name in file_names:
        print(f"--- 파일 로드 시도: {file_name} ---")
        try:
            df = pd.read_csv(file_name)

            if FILE_COLUMN not in df.columns or SCORE_COLUMN not in df.columns:
                print(f"경고: {file_name} 파일에 필요한 컬럼('{FILE_COLUMN}', '{SCORE_COLUMN}')이 없습니다. 건너뜁니다.")

                if 'Style_Similarity_Score' in df.columns:
                    print(f"참고: 이 파일은 'Style_Similarity_Score' 컬럼을 사용합니다. 코드 수정이 필요할 수 있습니다.")
                continue

            df = df[~df[FILE_COLUMN].isin(['---', 'Average Score', ''])].copy()
            df[SCORE_COLUMN] = pd.to_numeric(
                df[SCORE_COLUMN],
                errors='coerce'
            )

            df.dropna(subset=[SCORE_COLUMN], inplace=True)

            all_data.append(df)
            found_files_count += 1
            print(f"성공: {file_name} 파일의 {len(df)}개 항목을 수집했습니다.")

        except FileNotFoundError:
            print(f"오류: {file_name} 파일을 찾을 수 없습니다. 파일을 업로드해 주세요.")
        except Exception as e:
            print(f"오류: {file_name} 파일 처리 중 예상치 못한 오류 발생: {e}")

    if found_files_count == 0:
        print("\n오류: 처리할 유효한 데이터가 없어 평균 계산을 중단합니다.")
        return pd.DataFrame()
    elif found_files_count < len(file_names):
        print(f"\n경고: 요청하신 {len(file_names)}개 중 {found_files_count}개의 파일만 사용하여 평균을 계산합니다.")

    combined_df = pd.concat(all_data, ignore_index=True)
    average_scores_df = combined_df.groupby(FILE_COLUMN)[SCORE_COLUMN].mean().reset_index()
    average_scores_df.columns = [FILE_COLUMN, 'Average_Content_Similarity_Score']
    average_scores_df['Average_Content_Similarity_Score'] = average_scores_df['Average_Content_Similarity_Score'].apply(lambda x: f'{x:.4f}')
    average_scores_df.sort_values(by=FILE_COLUMN, inplace=True)

    return average_scores_df

FILE_NAMES = [
    "content_osaka_output.csv",
    # "content_camera_output.csv",
    "content_flower_output.csv",
    "content_yunah_output.csv"
]

OUTPUT_FILE_NAME = "average_content_similarity.csv"

final_average_df = calculate_average_scores(FILE_NAMES)

if not final_average_df.empty:
    final_average_df.to_csv(OUTPUT_FILE_NAME, index=False)
    print(f"최종 평균 CSV 파일 저장 완료: {OUTPUT_FILE_NAME}")

    print("\n[최종 평균 데이터 (상위 5개 항목)]")
    print(final_average_df.head().to_markdown(index=False, numalign="left", stralign="left"))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import os
import matplotlib.colors

STYLE_CSV_PATH = 'average_style_similarity.csv'
CONTENT_CSV_PATH = 'average_content_similarity.csv'
METHOD_NAME = 'Our Method'


def load_and_merge_scores(style_csv, content_csv):
    """
    두 CSV 파일을 File_Name을 기준으로 읽어와 병합하고 개별 데이터만 반환합니다.
    """

    # 파일 존재 여부 확인 (경로 '/content/'는 현재 작업 디렉토리로 간주하고 수정)
    if not os.path.exists(style_csv):
        print(f"오류: 스타일 CSV 파일이 존재하지 않습니다: {style_csv}")
        return None
    if not os.path.exists(content_csv):
        print(f"오류: 내용 CSV 파일이 존재하지 않습니다: {content_csv}")
        return None

    try:
        df_style = pd.read_csv(style_csv)
        df_content = pd.read_csv(content_csv)

        df_style = df_style[df_style.iloc[:, 0] != 'Average Score'].copy()
        df_content = df_content[df_content.iloc[:, 0] != 'Average Score'].copy()
        df_style = df_style[df_style.iloc[:, 0] != '---'].copy()
        df_content = df_content[df_content.iloc[:, 0] != '---'].copy()

        style_col = df_style.columns[1]
        content_col = df_content.columns[1]

        def clean_score(s):
            return s.astype(str).str.replace(' ', '').replace('---', np.nan).astype(float)

        df_style[style_col] = clean_score(df_style[style_col])
        df_content[content_col] = clean_score(df_content[content_col])

        merged_df = pd.merge(df_style, df_content, on=df_style.columns[0], how='inner')
        merged_df.rename(columns={merged_df.columns[0]: 'File_Name'}, inplace=True)


        if merged_df.empty or merged_df[style_col].isnull().all() or merged_df[content_col].isnull().all():
            print("오류: 병합 후 유효한 데이터 쌍이 없거나 점수가 모두 누락되었습니다.")
            return None

        style_scores = merged_df[style_col].tolist()
        content_scores = merged_df[content_col].tolist()
        file_names = merged_df['File_Name'].tolist()

        return style_scores, content_scores, file_names

    except Exception as e:
        print(f"오류: CSV 데이터 병합 또는 처리 중 문제가 발생했습니다: {e}")
        return None

result = load_and_merge_scores(STYLE_CSV_PATH, CONTENT_CSV_PATH)

if result is not None:
    style_scores, content_scores, file_names = result

    IDEAL_X, IDEAL_Y = 1.0, 1.0

    unique_file_names = sorted(list(set(file_names)))
    num_files = len(unique_file_names)

    cmap = plt.cm.get_cmap('tab20')
    colors = [cmap(i) for i in np.linspace(0, 1, num_files)]

    color_map = {name: colors[i] for i, name in enumerate(unique_file_names)}

    plt.figure(figsize=(14, 10))

    plt.scatter(IDEAL_X, IDEAL_Y, color='green', s=150, zorder=5)
    plt.text(IDEAL_X + 0.005, IDEAL_Y - 0.02, 'Ideal', color='green', fontsize=12)

    grouped_data = {}
    for style, content, fname in zip(style_scores, content_scores, file_names):
        if fname not in grouped_data:
            grouped_data[fname] = {'style': [], 'content': []}
        grouped_data[fname]['style'].append(style)
        grouped_data[fname]['content'].append(content)

    for fname, data in grouped_data.items():
        color = color_map[fname]

        label_name = os.path.splitext(fname)[0]

        plt.scatter(data['style'], data['content'],
                    color=color,
                    marker='o',
                    s=80,
                    alpha=0.8,
                    label=label_name,
                    zorder=6)

        for style, content in zip(data['style'], data['content']):
             plt.text(style + 0.005, content + 0.005, label_name,
                      fontsize=10,
                      color='black',
                      alpha=0.7,
                      zorder=7)

    plt.title(f'Individual Image Style vs. Content Similarity Trade-off ({METHOD_NAME})', fontsize=16)
    plt.xlabel('Style Similarity', fontsize=14)
    plt.ylabel('Content Similarity', fontsize=14)

    MIN_Y = 0.0
    MIN_X = 0.0

    plt.xlim(MIN_X, 1.05)
    plt.ylim(MIN_Y, 1.05)

    plt.grid(True, linestyle='-', alpha=0.6)

    plt.yticks(np.arange(MIN_Y, 1.01, 0.1))
    plt.xticks(np.arange(MIN_X, 1.01, 0.1))
    plt.legend(loc='lower left', title='Image File (Post-processed)', fontsize=10, title_fontsize=12)
    plt.savefig('trade_off_plot_colored.png')

    print("\n--- 파일별 색상 정보 ---")
    for name, color in color_map.items():

        hex_color = matplotlib.colors.rgb2hex(color[:3])
        print(f"파일: {os.path.splitext(name)[0]} -> 색상: {hex_color}")
    print("------------------------")

    print(f"\n총 {len(style_scores)}개의 파일 쌍에 대한 개별 점수 그래프가 'trade_off_plot_colored.png'로 저장되었습니다.")
else:
    print("\n그래프를 생성하지 못했습니다. CSV 파일 경로와 내용을 확인해 주세요.")
```
